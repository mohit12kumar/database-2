```python
import pandas as pd

import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
import matplotlib.pyplot as plt

file_path = r"https://raw.githubusercontent.com/mohit12kumar/database-2/refs/heads/main/kidney_disease.csv"

# Try different encodings
data = pd.read_csv(file_path, encoding="latin1")  # Alternative: "ISO-8859-1" or "utf-16"
print(data.head())  # Show first few rows
```

```python
le = LabelEncoder()

for col in data.columns:
    if data[col].dtype == 'object':
        data[col] = le.fit_transform(data[col])
```

```python
import seaborn as sns
corr = data.corr()
plt.figure(figsize=(25,25))
sns.heatmap(corr, annot=True, )
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# correlation matrix
corr = data.corr(numeric_only=True)

plt.figure(figsize=(15,25))
sns.heatmap(corr, annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")
plt.show()
```

```python
important_features = ['hemo', 'pcv', 'rc', 'sg',
    'al', 'htn', 'dm',
    'bgr', 'bu', 'sc',
    'appet', 'pe', 'ane']
x = data[important_features]
y =data["classification"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state=42)
svm_model = RandomForestClassifier()
svm_model.fit(x_train,y_train)
y_predict = svm_model.predict(x_test)
print(y_predict,y_test)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# Generate the confusion matrix
c1 = confusion_matrix(y_test, y_predict)
print("Confusion matrix:\n", c1)

# Combine true and predicted values to get all unique labels
unique_labels = sorted(list(set(y_test).union(set(y_predict))))

# Plot the confusion matrix
cm_display = ConfusionMatrixDisplay(confusion_matrix=c1, display_labels=unique_labels)
cm_display.plot(cmap='Blues', xticks_rotation=45)  # Rotate x-tick labels for better visibility
plt.show()
```

```python
print("\n test accuracy = ",accuracy_score(y_test,y_predict))
```
